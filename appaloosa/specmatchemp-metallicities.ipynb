```python
import pandas as pd
from pylab import *
import specmatchemp.library
import specmatchemp.plots as smplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lib = specmatchemp.library.read_hdf(wavlim=[5140,5200])
```

```python
fig = figure()
plt.plot(lib.library_params.Teff, lib.library_params.feh,'k.',)
plt.xlabel('T')
plt.ylabel('[Fe/H]')
```

```python

from matplotlib import rc
rc('text', usetex=True)
rc('font', size=13)
rc('legend', fontsize=13)
rc('text.latex', preamble=r'\usepackage{cmbright}')

fig = figure(figsize=(6,4))
t=lib.library_params.Teff[np.abs(lib.library_params.feh)<1]
m=lib.library_params.feh[np.abs(lib.library_params.feh)<1]
plt.scatter(t,m,s=10,c='k')
plt.xlabel(r'$T_{eff}$ (K)')
plt.ylabel('[Fe/H]')
plt.savefig('/home/ekaterina/Documents/Flares_in_Clusters_Paper/pic-dump/specmatchemp_feh.jpg', dpi=300)
```
