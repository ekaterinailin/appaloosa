```python
import pandas as pd
mine = pd.read_csv('/home/ekaterina/Documents/appaloosa/stars_shortlist/Pleiades/Pleiades_newmatch_no_outliers.txt',
                  header=None,names=['id','EPIC'])
theirs = pd.read_csv('/home/ekaterina/Documents/appaloosa/stars_shortlist/Pleiades/K2Campaign4targets.csv')
mine = mine.drop('id',axis=1)
#theirs = theirs['EPIC ID']
#theirs = theirs.rename(columns={'EPIC ID':'EPIC'})
theirs.head()
mine = mine.merge(theirs, how='inner', left_on='EPIC', right_on='EPIC ID')
invest = mine['Investigation IDs']
investclean = []
for k, item in invest.iteritems():
    iid,i0 = [], 0
    for i,letter in enumerate(item):
        if letter=='|':
            iid.append(item[i0:i].strip())
            i0 = i + 1
    iid.append(item[i0:].strip())
    invest[k] = iid
    
mine['GO'] = invest
print(mine.head())
mine2 = pd.DataFrame()
for j, row in mine.iterrows():
    for go in row['GO']:
        new = row
        new.GO = go
        mine2 = mine2.append(new,ignore_index=True)
```
