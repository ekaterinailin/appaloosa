```python

import numpy as np
import specmatchemp.library
import matplotlib.pyplot as plt
%matplotlib inline

lib = specmatchemp.library.read_hdf(wavlim=[5140,5200])
```

```python
fig = plt.figure(figsize=(7,5))
t=lib.library_params.Teff[np.abs(lib.library_params.feh)<1]
m=lib.library_params.feh[np.abs(lib.library_params.feh)<1]
plt.scatter(t,m,s=10,c='k')
plt.xlabel('Effective Temperature (K)')
plt.ylabel('[Fe/H]')
#plt.savefig('/home/ekaterina/Documents/Flares_in_Clusters_Paper/pic-dump/specmatchemp_feh.jpg', dpi=300)
```
